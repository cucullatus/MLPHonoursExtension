```python
import matplotlib.pyplot as plt
idim = 10
rng = numpy.random.RandomState([2015, 1, 1])

odim = 10
x = rng.uniform(-2, 2, (20, idim))
s = ComplexAbs(idim=idim, odim=odim, rng=rng)
h = s.fprop(x)
plt.matshow(1./h)
plt.show()
def f(x):
    return numpy.sum(s.fprop(x)), 0
plt.matshow(numerical_gradient(f=f,x=x,eps=5))
plt.show()
```
